#### 문제
비어있는 공집합 S가 주어졌을 때, 아래 연산을 수행하는 프로그램을 작성하시오.

add x: S에 x를 추가한다. (1 ≤ x ≤ 20) S에 x가 이미 있는 경우에는 연산을 무시한다.

remove x: S에서 x를 제거한다. (1 ≤ x ≤ 20) S에 x가 없는 경우에는 연산을 무시한다.

check x: S에 x가 있으면 1을, 없으면 0을 출력한다. (1 ≤ x ≤ 20)

toggle x: S에 x가 있으면 x를 제거하고, 없으면 x를 추가한다. (1 ≤ x ≤ 20)

all: S를 {1, 2, ..., 20} 으로 바꾼다.

empty: S를 공집합으로 바꾼다. 

#### 입력

첫째 줄에 수행해야 하는 연산의 수 M (1 ≤ M ≤ 3,000,000)이 주어진다.

둘째 줄부터 M개의 줄에 수행해야 하는 연산이 한 줄에 하나씩 주어진다.

#### 출력
check 연산이 주어질때마다, 결과를 출력한다.

```python
m = int(input())
s = []
prnt = []
for _ in range(m):
    operations = input().split()
    op = operations[0]
    if op == 'add':
        x = int(operations[1])
        if x not in s:
            s.append(x)
    elif op == 'remove':
        x = int(operations[1])
        if x in s:
            s.remove(x)
    elif op == 'check':
        x = int(operations[1])
        if x in s:
            prnt.append(1)
        else:
            prnt.append(0)
    elif op == 'toggle':
        x = int(operations[1])
        if x in s:
            s.remove(x)
        else:
            s.append(x)
    elif op == 'all':
        s = list(range(1, 21))
    elif op == 'empty':
        s = []

for rst in prnt:
    print(rst)
```

```python
m = int(input())
s = {i+1:0 for i in range(20)}
prnt = []
for _ in range(m):
    operations = input().split()
    op = operations[0]
    if op == 'add':
        x = int(operations[1])
        if s[x] == 0:
            s[x] = 1
    elif op == 'remove':
        x = int(operations[1])
        if s[x] == 1:
            s[x] = 0
    elif op == 'check':
        x = int(operations[1])
        if s[x] == 1:
            prnt.append(1)
        else:
            prnt.append(0)
    elif op == 'toggle':
        x = int(operations[1])
        if s[x] == 1:
            s[x] = 0
        else:
            s[x] = 1
    elif op == 'all':
        s = {i+1:1 for i in range(20)}
    elif op == 'empty':
        s = {i+1:0 for i in range(20)}

for rst in prnt:
    print(rst)
```

```python
m = int(input())
s = set()
prnt = []
for _ in range(m):
    operations = input().split()
    op = operations[0]
    if op == 'add':
        x = int(operations[1])
        s.add(x)
    elif op == 'remove':
        x = int(operations[1])
        s.discard(x)
    elif op == 'check':
        x = int(operations[1])
        if x in s:
            prnt.append(1)
        else:
            prnt.append(0)
    elif op == 'toggle':
        x = int(operations[1])
        if x in s:
            s.remove(x)
        else:
            s.add(x)
    elif op == 'all':
        s = set(range(1, 21))
    elif op == 'empty':
        s = set()

for rst in prnt:
    print(rst)
```
