#### 문제
트럭을 타고 이동하던 상근이는 경찰의 검문을 받게 되었다. 경찰은 상근이가 운반하던 화물을 하나하나 모두 확인할 것이기 때문에, 검문하는데 엄청나게 오랜 시간이 걸린다.

상근이는 시간을 때우기 위해서 수학 게임을 하기로 했다.

먼저 근처에 보이는 숫자 N개를 종이에 적는다. 그 다음, 종이에 적은 수를 M으로 나누었을 때, 나머지가 모두 같게 되는 M을 모두 찾으려고 한다. M은 1보다 커야 한다.

N개의 수가 주어졌을 때, 가능한 M을 모두 찾는 프로그램을 작성하시오.

#### 입력
첫째 줄에 종이에 적은 수의 개수 N이 주어진다. (2 ≤ N ≤ 100)

다음 줄부터 N개 줄에는 종이에 적은 수가 하나씩 주어진다. 이 수는 모두 1보다 크거나 같고, 1,000,000,000보다 작거나 같은 자연수이다. 같은 수가 두 번 이상 주어지지 않는다.

항상 M이 하나 이상 존재하는 경우만 입력으로 주어진다.

#### 출력
첫째 줄에 가능한 M을 공백으로 구분하여 모두 출력한다. 이때, M은 증가하는 순서이어야 한다.

#### 예제 입력 1 
3

6

34

38

#### 예제 출력 1 
2 4

#### 예제 입력 2 
5

5

17

23

14

83

#### 예제 출력 2 
3

```python
from math import gcd
import sys

n = int(sys.stdin.readline())
n_ls = sorted([int(sys.stdin.readline()) for _ in range(n)])

diff = [n_ls[i] - n_ls[i - 1] for i in range(1, n)]
g = diff[0]
for d in diff[1:]:
    g = gcd(g, d)
    
result = []
for i in range(2, g + 1):
    if g % i == 0:
        result.append(str(i))
        
print(' '.join(result))
```

```python
from math import sqrt, gcd

def deli(num):
    ls = []
    for i in range(1, int(sqrt(num)) + 1):
        if num % i == 0:
            if i != 1:
                ls.append(i)
            if n // i != i:
                ls.append(num // i)
            
    return ls
```

```python
n = int(input())
n_ls = []
min_num = int(1e9)
diff_num = int(1e9)
results = []
for _ in range(n):
    num = int(input())
    
    for nb in n_ls:
        diff = abs(nb - num)
        if diff_num > diff:
            diff_num = diff
            
    n_ls.append(num)
    
    if min_num > num:
        min_num = num 
        
for i in range(min(min_num, diff_num) + 1):   
    gcd_num = n_ls[0] - i 
    for j in range(1,n):
        gcd_num = gcd(gcd_num, n_ls[j] - i)
    if gcd_num != 1:
        m_ls = deli(gcd_num)
        results.extend(m_ls)

for rst in sorted(set(results)):
    print(rst, end=' ')
```
