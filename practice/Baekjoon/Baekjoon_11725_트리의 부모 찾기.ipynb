#### 문제
루트 없는 트리가 주어진다. 이때, 트리의 루트를 1이라고 정했을 때, 각 노드의 부모를 구하는 프로그램을 작성하시오.

#### 입력
첫째 줄에 노드의 개수 N (2 ≤ N ≤ 100,000)이 주어진다. 둘째 줄부터 N-1개의 줄에 트리 상에서 연결된 두 정점이 주어진다.

#### 출력
첫째 줄부터 N-1개의 줄에 각 노드의 부모 노드 번호를 2번 노드부터 순서대로 출력한다.

```python
def dfs(num, conn):
    for c in conn[num]:
        if visited[c] == -1:
            visited[c] = num
            dfs(c, conn)
```

```python
n = int(input())
conn = {i+1: [] for i in range(n)}
visited = [-1 for _ in range(n+1)]
for _ in range(n-1):
    a, b = map(int, input().split())
    conn[a].append(b)
    conn[b].append(a)
    
visited[1] = 0
dfs(1, conn)

for i in visited[2:]:
    print(i)
```

```python
from collections import deque


n = int(input())
conn = {i+1: [] for i in range(n)}
visited = [-1 for _ in range(n+1)]
for _ in range(n-1):
    a, b = map(int, input().split())
    conn[a].append(b)
    conn[b].append(a)
    
visited[1] = 0
q = deque()
q.append(1)
while q:
    parent = q.popleft()
    for i in conn[parent]:
        if visited[i] == -1:
            visited[i] = parent
            q.append(i)
            

for i in visited[2:]:
    print(i)
```
